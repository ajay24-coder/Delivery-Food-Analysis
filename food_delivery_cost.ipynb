#cleaning

```python
#cleaning data
df["Delivery Date and Time"] = pd.to_datetime(df["Delivery Date and Time"]) # Use pd.to_datetime for datetime conversion
df.info()
```

```python
a = "15% New User"
z = a.split()
z[0]
```

```python
def extract(value):
  a =str(value).split(" ")
  return a[0]

df["Discounts and Offers"] = df["Discounts and Offers"].apply(extract)
df.head()
```

```python
def removea(value):
  if isinstance(value, str) and "%" in value:  # Check if value is a string and contains '%'
    a = value.replace("%","")
    return float(a)
  elif isinstance(value, str):  # Handle string values without '%'
    return float(value)
  else:
    return value  # Return original value if it's not a string (e.g., already a float)

df["Discounts and Offers"] = df["Discounts and Offers"].apply(removea)
df.head()
```

```python
df.loc[(df["Discounts and Offers"] <=15),"Discounts and Offers"]=(df["Discounts and Offers"]/100)* df["Order Value"]
df.head()
```

```python
df["Discounts and Offers"] = df["Discounts and Offers"]
df.head()
```

```python
df["Costs"]= df["Delivery Fee"] + df["Discounts and Offers"] + df["Payment Processing Fee"]
df.head()
```

```python
# Calculate profit (assuming 'Order Value' and 'Costs' columns exist)
df["Profit"] = df["Order Value"] - df["Costs"]

# Now you can calculate the sum of profits
df["Profit"].sum()
```

```python
cost_dist = df[["Delivery Fee", "Payment Processing Fee", "Discounts and Offers"]].sum()
cost_dist
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (3,3))
plt.pie(cost_dist,labels = cost_dist.index ,autopct="%1.3f%%")
plt.title("Cost Distribution")
plt.show()
```
